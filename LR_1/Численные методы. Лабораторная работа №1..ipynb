Находит правильно, если учитывать погрешностии округлять результат (например с $\epsilon$ = $10^{-16}$)

```python
def mysin(x):
    u = x
    i = 1
    flag = False
    summ = u
    while True:
        u = u * (- (x ** 2) / ((2*(i-1)+2)*(2*(i-1)+3)))
        i += 1
        summ += u
        if flag:
            break
        if np.abs(u) < 1e-17:
            flag = True
    return summ

points = [0, np.pi / 3, np.pi / 2, np.pi, np.pi*2]
sins = [mysin(x) for x in points]
print(points)
print(sins)
```
